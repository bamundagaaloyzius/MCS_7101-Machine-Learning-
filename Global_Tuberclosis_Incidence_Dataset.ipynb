How do we know the features in the dataset?

The list(data.columns) function creates a list of the elements in the first row thus the features.
Lets consider Global_Incidence and Africa datasets below.

```python
list(Global_Incidence.columns)
```

- **IND_ID** : Indicator unique identifier
- **IND_CODE** : Indicator code
- **IND_UUID** : Indicator code
- **IND_PER_CODE** : Indicator code
- **DIM_TIME** : The year of submission
- **DIM_TIME_TYPE** : The time allocation
- **DIM_GEO_CODE_M49** : The geographical code
- **DIM_GEO_CODE_TYPE** : The geographical entity
- **DIM_PUBLISH_STATE_CODE** : This states whether the data is published by WHO or NOt.
- **IND_NAME**: Disease case for this case Tuberclosis
- **GEO_NAME_SHORT** : country or region of data collection
- **RATE_PER_100000_N** : Average number of incidence cases for every 100,000 people for that year.
- **RATE_PER_100000_NL** : Maximum number of incidence cases for every 100,000 people for that year.
- **RATE_PER_100000_NU** : Minimum number of incidence cases for every 100,000 people for that year.

What is the incidence trend for Low income countries?

The incidence trend over the period 2000 to 2022 can be acquired by using the .loc function in the GEO_NAME_SHORT column to look for 'Low-income-countries'. The incidence drops from an average of 274 to 181 cases per 100,000 people.

```python
Income.loc[Income['GEO_NAME_SHORT'] == 'Low-income economies']
```
